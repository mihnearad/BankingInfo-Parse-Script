```python
import pandas as pd

df_rev=pd.read_csv('.data/REVBUS.csv')
```

```python
df_rev.head()
```

```python
new_order_column=['Date started (UTC)','Description','Reference','Balance','Fee currency']

df_rev=df_rev[new_order_column]
df_rev.head()
```

```python
df_rev["Date started (UTC)"] = (
    pd.to_datetime(df_rev["Date started (UTC)"])
    .dt.strftime("%d-%m-%Y")
    .sort_values(ascending=True)
)
```

```python
filtered_rev = df_rev[~df_rev["Description"].str.contains(r"Mihnea|RON")]
filtered_rev 
```
